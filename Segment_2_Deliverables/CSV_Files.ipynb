```python
# Read in FTB csv
ftb_df = pd.read_csv(ftb_path)
ftb_df.head()
```

```python
ftb_drop_df = ftb_df.drop(ftb_df.index[range(2)])
ftb_drop_df.head()
```

```python
# Use row #2 to rename as the new header row of the dataframe
reset_ftb_df = ftb_drop_df.rename(columns=ftb_drop_df.iloc[0]).drop(ftb_drop_df.index[0])
reset_ftb_df
```

```python
# Split the "QTR" column into one columns for Month and Year to better match formatting of other data source
reset_ftb_df['Year'] = reset_ftb_df['QTR'].replace(".$", "", regex=True)
reset_ftb_df.head()
```

```python
# Split the "Year" Column to remove . from year
reset_ftb_df[['Year','Del']] = reset_ftb_df['Year'].str.split('.', 1, expand=True)
reset_ftb_df.head()
```

```python
# Delete QTR and keep only CA, Riverside, LA, SD, SF and Year
filter_ftb_df = reset_ftb_df.filter(items=['Year', 'CA', 'LA', 'Riverside','San Diego', 'San Francisco'])
filter_ftb_df = filter_ftb_df.rename(columns={'LA':'Los Angeles'})
filter_ftb_df.head()
```

```python
# Get Column List Name
ftb_columns= list(filter_ftb_df.columns)
ftb_columns.remove('Year')
print(ftb_columns)
```

```python
filter_ftb_df.head(20)
filter2_ftb_df = filter_ftb_df.dropna()
```

```python
# Convert values to numeric numbers and create new dataframe
import numpy as np
filter2_ftb_df[ftb_columns] = filter2_ftb_df[ftb_columns].replace("%", "", regex=True)
num_ftb_df = filter2_ftb_df.astype(str).astype(int)
num_ftb_df = num_ftb_df.reset_index()
num_ftb_df.head()
```

```python
# Avg years with group by function
avg_ftb = num_ftb_df.groupby('Year')[ftb_columns].mean()
avg_ftb.head()
```

```python
# Convert Year from index to column
clean_avg_ftb = avg_ftb.reset_index()
clean_avg_ftb = clean_avg_ftb.rename(columns={'index':'Year'})
clean_avg_ftb.head()
```

```python
# Retrieve column list and remove year
ftb_columns = list(clean_avg_ftb.columns)
ftb_columns.remove('Year')
print(ftb_columns)
```

```python
# Transpose the data points into a vertical table
clean_ftb_df = clean_avg_ftb.melt(id_vars = ['Year'], value_vars = ftb_columns)
clean_ftb_df = clean_ftb_df.rename(columns={'variable':'County','value':'First_Time_Buyer_Rate'})
clean_ftb_df
```

```python
# Merge FTB dataframe with clean_final_df
clean_merge_df = clean_final_df.merge(clean_ftb_df)
clean_merge_df
```
