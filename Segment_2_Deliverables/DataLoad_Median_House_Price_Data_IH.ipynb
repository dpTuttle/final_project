```python
# Add pandas as dependency
import pandas as pd
```

```python
# Import File
median_house_price_data = "../Resources/MedianPricesofExistingDetachedHomesHistoricalData.csv"
```

```python
# Convert to DataFrame
median_house_df = pd.read_csv(median_house_price_data)
median_house_df.head(10)
```

```python
# Remove rows 1 - 5 that were notes columns
clean_median_house_df = median_house_df.drop(labels=range(0,6), axis=0)
clean_median_house_df
```

```python
# Use row #6 to rename as the new header row of the dataframe
clean2_median_house_df = clean_median_house_df.rename(columns=clean_median_house_df.iloc[0]).drop(clean_median_house_df.index[0])
clean2_median_house_df
```

```python
# Split the "Mon-Yr" column into two columns for Month and Year to better match formatting of other data source
clean2_median_house_df[['month','year']] = clean2_median_house_df['Mon-Yr'].str.split('-', 1, expand=True)
clean2_median_house_df
```

```python
# Drop all the columns with NaN values
clean3_median_house_df = clean2_median_house_df.dropna(axis = 'columns')
clean3_median_house_df
```

```python
# Drop the Mon-Year column
clean4_median_house_df = clean3_median_house_df.drop(columns=['Mon-Yr'])
clean4_median_house_df
```

```python
# Reset the index column to start from 1
clean5_median_house_df = clean4_median_house_df.reset_index()
clean5_median_house_df
```

```python
# Drop the "Index" column
clean6_median_house_df = clean5_median_house_df.drop(columns=['index'])
clean6_median_house_df
```

```python
# Create the column names into a list
column_list = list(clean6_median_house_df.columns)
column_list.remove("year")
column_list.remove("month")

column_list
```

```python
# Transpose the data points into a vertical table
clean7_median_house_df = clean6_median_house_df.melt(id_vars = ['year','month'], value_vars = column_list)
clean7_median_house_df
```
